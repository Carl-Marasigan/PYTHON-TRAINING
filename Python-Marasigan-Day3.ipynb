```python
from abc import abstractmethod


class Shape:
    def area(self):
        pass

class Polygon(Shape):
    def __init__(self, sides):
        self.sides = sides

    @abstractmethod
    def perimeter(self):
        pass

class Rectangle(Polygon):
    def __init__(self, width, height):
        self.width = width
        self.height = height
        super().__init__(4)

    def perimeter(self):
        return 2 * (self.width + self.height)

    def area(self):
        return self.width * self.height

class Triangle(Polygon):
    def __init__(self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
        super().__init__(3)

    def perimeter(self):
        return self.side1 + self.side2 + self.side3

class Square(Rectangle):
    def __init__(self, side_length):
        self.side_length = side_length
        super().__init__(side_length, side_length)

    def area(self):
        return self.side_length ** 2

    def perimeter(self):
        return 4 * self.side_length

# Example usages

rectangle = Rectangle(4, 6)
print("Rectangle area:", rectangle.area())
print("Rectangle perimeter:", rectangle.perimeter())

square = Square(4)
print("Square area:", square.area())
print("Square perimeter:", square.perimeter())

triangle = Triangle(3, 4, 5)
print("Triangle perimeter:", triangle.perimeter())
```

```python
import tkinter

# Create the main window
root = tkinter.Tk()
root.title("Organizing Widgets with Frames (Using StringVar)")

# Frame for the top section (contains a label and entry field)
top_frame = tkinter.Frame(root)
top_frame.pack(padx=10, pady=10, fill="x")

# Create a StringVar for the entry field
entry_var = tkinter.StringVar()

# Add a label and entry inside the top_frame
label = tkinter.Label(top_frame, text="Enter your name:")
label.pack(side="left", padx=5)

entry = tkinter.Entry(top_frame, textvariable=entry_var)
entry.pack(side="left", padx=5)

# Frame for the middle section (contains buttons)
button_frame = tkinter.Frame(root)
button_frame.pack(padx=10, pady=10, fill="x")

# Add buttons to button_frame
button1 = tkinter.Button(button_frame, text="Submit")
button1.pack(side="left", padx=5)

button2 = tkinter.Button(button_frame, text="Clear")
button2.pack(side="left", padx=5)

# Frame for the bottom section (contains a result label)
bottom_frame = tkinter.Frame(root)
bottom_frame.pack(padx=10, pady=10, fill="x")

# Add a result label inside bottom_frame
result_label = tkinter.Label(bottom_frame, text="Result will be shown here")
result_label.pack(side="left", padx=5)

# Function to handle the submit action
def submit_action():
    name = entry_var.get()  # Get the value from the StringVar
    result_label.config(text=f"Hello, {name}!")

# Function to handle the clear action
def clear_action():
    entry_var.set("")  # Clear the StringVar, which will update the entry widget
    result_label.config(text="Result will be shown here")

# Bind button actions to functions
button1.config(command=submit_action)
button2.config(command=clear_action)

# Start the application loop
root.mainloop()
```
