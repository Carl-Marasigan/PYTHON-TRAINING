```
# Activity 5: Simple Calculator

# FUNCTION CALCULATOR
def calculator(num1, num2,operator):
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        return num1 - num2
    elif operator == "*":
        return num1 * num2
    elif operator == "/":
        return num1 / num2
    else:
        return "Error: Invalid operator. Please use +, -, *, or /"


num1 = int(input("Enter first number: "))
num2= int(input("Enter second number: "))
operator = input("Enter operator (+, - , *, and /): ")
result = calculator(num1,num2,operator)

# Output
print(result)




```

```
#Activity 6:  Catch the Possible Error

def concatenate_strings(a, b):
    try:
        return a + b
    except TypeError:
        return "Error: Both input must be String."

def safe_divide(x, y):
    try:
        return x / y
    except ZeroDivisionError:
        return "Error: Cannot divide by zero."

def float_converter(number):
    try:
        return float(number)
    except ValueError:
        return "Error: Please enter a valid number."

#USER INPUTS

a = input("Enter a first string: ")
b = int(input("Enter a second string: "))
print(concatenate_strings(a,b))

x= float(input("Enter the first number: "))
y= float(input("Enter the second number: "))
print(safe_divide(x,y))

number = input("Enter a number to convert a float number: ")
print(float_converter(number))

# print(concatenate_strings("Age:", 25))
# print(safe_divide(10,0))
# print(float_converter("qwerty"))
```
