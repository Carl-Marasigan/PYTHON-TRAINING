```python
# Grouping
fruits = ["apple", "banana", "kiwi", "cherry", "lime", "avocado"]
print(fruits)
fruits_by_length = dict()
for fruit in fruits:
    length = len(fruit)
    if length not in fruits_by_length:
        fruits_by_length[length] = []

    fruits_by_length[length].append(fruit)

print(fruits_by_length)
```

```python
# Activity 2:  Christmas Playlist

def add(song, playlist):
    playlist.append(song)
    print("Song added")

def remove(song, playlist):
    for item in playlist:
        if item.lower() == song.lower():
            playlist.remove(item)
            print("Song removed")
            return
        print("Song not found")

def play(playlist):
    if playlist:
        print("Now Playing", playlist[0])
    else:
        print("Playlist empty")

def show_all(playlist):
    print("Christmas Playlist")
    for index, song in enumerate(playlist,start=1):
        print(f"{index}. {song}")

def playlist_app():
    playlist = ['Hawak mo ang Beat','Kumukutikutitap','Pasko Nananaman','Christmas in our hearts']

try:
    while True:
        print("\n=====CHOOSE OPTION=====")
        print("1. Add Song")
        print("2. Remove Song")
        print("3. Play Song")
        print("4. Show Playlist")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            song = input("Enter Song name: ")
            add(song, playlist)
            print(playlist)
        elif choice == "2":
            song = input("Enter Song name to remove: ")
            remove(song, playlist)
        elif choice == "3":
            play(playlist)
        elif choice == "4":
            show_all(playlist)
        elif choice == "5":
            print("Thank you for using this program!")
            break

        else:
            print("Invalid choice!")



except ValueError:
    print(f"Index is not valid index")

playlist_app()

```
